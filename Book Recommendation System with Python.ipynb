```python
#preprocesing
```

```python
#turn to numeric
```

```python
df = df[pd.to_numeric(df["average_rating"], errors="coerce").notnull()]
```

```python
df["average_rating"] = df["average_rating"].astype("float32")
```

```python
#create a new column authors + title because we use it like this
```

```python
df['book_content'] = df['title'] + ' ' + df['authors']
```

```python
#vektoryzing
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df['book_content'])
```

```python
#compute cosinus similatory
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
#define a function for recomandation
```

```python
def recommend_books(book_title, cosine_sim=cosine_sim):
    idx = df[df['title'] == book_title].index[0]
    sim_scores = sorted(enumerate(cosine_sim[idx]), key=lambda x: x[1], reverse=True)[1:11]
    return df['title'].iloc[[book[0] for book in sim_scores]]
```

```python
# example usage
book_title = "The Catcher in the Rye"
recommended_books = recommend_books(book_title)
print(recommended_books)
```
